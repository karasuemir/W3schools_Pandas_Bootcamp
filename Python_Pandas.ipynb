                                                                                                                        Pandas Series

    Seris nedir?
# Pandas Serisi, tablodaki bir sütun gibidir.

# Herhangi bir türden veriyi tutan tek boyutlu bir dizidir.

```python
#Series oluşturma:

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)
```

```python
# series im ilk değerini döndürür.
print(myvar[0])
```

```python
a = [1, 7, 2]
# index parametresi ile argümanların indexlerini adlandırabilirsiniz.
myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)
```

```python
# sözlük biçiminde bir series oluşturursanız anahtarlar seriesin etiketlei haline gelir.
calories = {"day1": 420, "day2": 380, "day3": 390}

# index parametresini kullanarak sözlükteki belirli anahtarları çekerekte series oluşturabiliriz.(sözlükteki bütün değişkenleri kullanmaya biliriz)
myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
```

                    Boş Hücreler

# Boş hücreler, verileri analiz ettiğinizde potansiyel olarak size yanlış bir sonuç verebilir.

    Satırları Kaldır
# Boş hücrelerle başa çıkmanın bir yolu, boş hücreler içeren satırları kaldırmaktır.

# Bu genellikle iyidir, çünkü veri kümeleri çok büyük olabilir ve birkaç satırın kaldırılması sonuç üzerinde büyük bir etki yaratmaz.

```python
# dropna() fonksiyonu birlikte kullanılan df nin içindeki boş hücreleri ortadan kaldırır ve değişkene atar.
new_df = df.dropna()

# new_df fonksiyonu uzun olduğundan head() fonksiyonunu kullandım.
new_df.head()
```

```python
# inplance = True parametresini girdiğimizden df deki tüm NULL gözlemlerini kaldırır.
df.dropna(inplace = True)

# df fonksiyonu uzun olduğundan head() fonksiyonunu kullandım.
df.head()
```

     Boş değerleri kaldırma

```python
df = pd.read_csv('data.csv')
```

```python
# fillna fonksiyonu dfdeki boş gözlemlerin yerine belirtilen değeri girer.
df.fillna(130, inplace = True)
```

```python
# filina fonksiyonunu sadece belirtilen sutun(değişkenler) içinde yapabiliriz.
df["Calories"].fillna(130, inplace = True)
```

    Ortalama, Medyan veya Mod Kullanarak Değiştir

```python
df = pd.read_csv('data.csv')

# df dataframeindeki Calories sutunundaki(değişkenindeki) gözlemlerin ortalamalarını alıyoruz.
x = df["Calories"].mean()
print(x)

# filina fonksiyonu ile df dataframedeki boş yerlere ortalama değerimizi atıyoruz.
df["Calories"].fillna(x, inplace = True)
df.head(20)
```

```python
df = pd.read_csv('data.csv')

# df dataframeindeki Calories sutunundaki(değişkenindeki) gözlemlerin medianlarını alıyoruz.
x = df["Calories"].median()
x
# filina fonksiyonu ile df dataframedeki boş yerlere median değerimizi atıyoruz.
df["Calories"].fillna(x, inplace = True)
df.head(20)
```

```python
df = pd.read_csv('data.csv')

# df dataframeindeki Calories sutunundaki(değişkenindeki) gözlemlerin modunu alıyoruz.
x = df["Calories"].mode()[0]
x

# filina fonksiyonu ile df dataframedeki boş yerlere mod değerimizi atıyoruz.
df["Calories"].fillna(x, inplace = True)
df.head(20)
```
